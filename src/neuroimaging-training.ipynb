## Dataset

To run this tutorial, please, download the following data (link below) and put the patients' folders as well as merged patient_info tables into the 'data' folder of this repo:  
http://lit.fe.uni-lj.si/tools.php  
Lesjak, Žiga, Franjo Pernuš, Boštjan Likar, and Žiga Špiclin. "A Novel Public MR Image Dataset of
Multiple Sclerosis Patients With Lesion Segmentations Based on Multi-rater Consensus",
Neuroinformatics (2017), DOI: 10.1007/s12021-017-9348-7  

This dataset contains brain MRI of multiple sclerosis (MS) patients. The MRI sequences are typical for MS: 
* T1w - for basic anatomical scans, 
* T1w contrast enhanced - after gadolinium injection, shows vascular structures and breakdowns in blood-brain barrier, 
* T2w - to visualise white matter hyperintensities == brain lesions, 
* FLAIR -  to visualise white matter hyperintensities == brain lesions + also shows anatomical structures.  

The following binary masks are available:
* brainmask,
* lesionmask.
  
Full data description could be found via the same link.

* DICOM - 'industrial' format: https://www.dicomstandard.org/

## Lesions quantification

The dataset contains data for only 30 patients, which is a fairly small cohort. The EDSS scores are available for the patients. We will try to link lesion load (total lesion volume and lesion count) with EDSS.

```python
def les_volume(les_mask_nii):
    # function to calculate a total volume of lesions in brain in mm^3
    # based on lesion mask nifti object
    pixdim = lesionmask_nii.header['pixdim']
    vox_size = pixdim[1]*pixdim[1]*pixdim[3]
    les_mask = les_mask_nii.get_fdata()
    les_load = np.sum(les_mask)
    return les_load*vox_size
```

```python
def les_count(les_mask):
    # function to calculate a total number of lesions
    # based on binary lesion mask
    les_mask_labeled = label(les_mask)
    return np.max(les_mask_labeled)
```

Reading the patient info file, containing EDSS.

```python
patient_info = pd.read_csv('data/patient_info.csv')
```

Adding lesion load data into the dataframe

```python
patient_info['les_volume'] = 0
patient_info['les_count'] = 0
for index, row in patient_info.iterrows():
    patient_id = str(row['patient_id']).zfill(2)
    lesionmask_path = '../data/patient' + patient_id + '/patient' + patient_id + '_consensus_gt.nii.gz'
    lesionmask_nii = nib.load(lesionmask_path)
    lesionmask = lesionmask_nii.get_fdata()
    l_vol = les_volume(lesionmask_nii)
    l_count = les_count(lesionmask)
    patient_info.loc[index, 'les_volume'] = l_vol
    patient_info.loc[index, 'les_count'] = l_count
```

Demographic and clinical information, and imaging features

```python
patient_info
```

Spearman correlation between the imaging features and EDSS

```python
r_edss_vol = patient_info['edss'].corr(patient_info['les_volume'], method = 'spearman')
r_edss_count = patient_info['edss'].corr(patient_info['les_count'], method = 'spearman')

print ('r_EDSS_volume = ', r_edss_vol)
print ('r_EDSS_count = ', r_edss_count)
```

We did not find any significant correlation, which could be due to the small sample size. Anyway, we can recommend to try the other imaging features. The promising ones can be related to lesions location.
